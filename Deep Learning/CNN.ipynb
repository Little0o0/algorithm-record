```python
import torch
import torch.nn as nn
```

```python
class CNN(nn.Module):
    def __init__(self, output_dim):
        super(CNN, self).__init__()
        
        # m, 3, 64, 64 -> m, 64, 62, 62 ->  m, 64, 31, 31 ->  m, 128, 29, 29
        self.conv1 = nn.Conv2d(3, 64, 3, 1)
        self.bn = nn.BatchNorm2d(64)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(2, stride = 2) # attention! Default value is kernel_size
        self.conv2 =  nn.Conv2d(64, 128, 3, 1)
        self.fc1 = nn.Linear(128 * 29 * 29, 1024)
        self.fc2 = nn.Linear(1024, output_dim)
    
    def forward(self, x):
        x = self.conv1(x)
        x = self.bn(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = torch.flatten(x, 1) # attention here !!!!
        x = self.fc1(x)
        x = self.fc2(x)
        return x
```

```python
x = torch.rand((64, 3, 64, 64))
model = CNN(output_dim = 10)
model(x)
```
